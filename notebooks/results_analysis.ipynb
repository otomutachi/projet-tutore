# Analyse des résultats

Ce notebook charge les résultats de l’évaluation et affiche un résumé simple.

```python
import json
from pathlib import Path
import pandas as pd

results_path = Path('../data/results/evaluation_results.json')
with open(results_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

df = pd.DataFrame(data.get('results', []))
df.head()
```

```python
if not df.empty:
    print('Nombre dévaluations :', len(df))
    print('Score moyen combiné :', round(df['combined_score'].mean(), 3))
    print('Score moyen de correction :', round(df['correctness_score'].mean(), 3))
    print('Score moyen de sécurité :', round(df['security_score'].mean(), 3))
    print(df.groupby('llm_provider')['combined_score'].mean())
```
